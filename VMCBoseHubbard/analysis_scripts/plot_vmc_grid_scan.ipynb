```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Explicitly assign column names
col_names = ["U", "kappa", "energy", "sem"]

# Read the data, skipping comments and assigning column labels
df = pd.read_csv("vmc_grid_scan.dat", 
                 delim_whitespace=True, 
                 comment="#", 
                 names=col_names)

print(df.head())
```

```python
plt.figure(figsize=(10, 6))
for U in sorted(df["U"].unique()):
    subset = df[df["U"] == U]
    plt.errorbar(subset["kappa"], subset["energy"], yerr=subset["sem"], label=f"U = {U}", fmt='o')

plt.axvline(1.4)
plt.xlabel("κ")
plt.ylabel("Energy")
plt.title("VMC Energy vs κ for different U")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
